```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from sklearn import linear_model
from pandas.tools.plotting import scatter_matrix
import statsmodels.api as sm
import statsmodels.stats.api as sms
import statsmodels.formula.api as ols
from sklearn import preprocessing
matplotlib.style.use('ggplot')
import math
```

```python
tree_data =  pd.read_csv('Ex02_TreeData_Regreesion.csv')
```

```python
names = tree_data.columns.tolist()
x = tree_data.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(tree_data)
tree_data = pd.DataFrame(x_scaled)
tree_data.columns = names
tree_data = tree_data[tree_data.Nt.values != 0]
```

```python
splitend = int(0.8 * len(tree_data))
tree_data_shuffled = tree_data.sample(frac=1)
train, test = tree_data_shuffled[:splitend], tree_data_shuffled[splitend:]
```

```python
correlation_before_feature_selection =  tree_data.corr()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlation_before_feature_selection, vmin = -1, vmax = 1)
fig.colorbar(cax)
ticks = np.arange(0, len(names), 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```python
scatter_matrix(tree_data, alpha=0.2, figsize=(10, 10), diagonal='kde')
plt.show()
```

```python
X_train = train.iloc[:, 0:4]
Y_train = train.iloc[:, 4] 
X_test = test.iloc[:, 0:4]
Y_test = test.iloc[:, 4]
```

```python
model_scaled = ols.ols(formula='Age ~ np.log(Nt)', data=train).fit()
print (model_scaled.summary()) 
res = model_scaled.predict(X_test)
plt.plot(list(Y_test), list(res), color='Red')
plt.show()
```

```python
residual = res - Y_test
residual.plot(kind= 'box')
plt.show()
```

```python
np.sqrt(np.mean(residual ** 2))
```
