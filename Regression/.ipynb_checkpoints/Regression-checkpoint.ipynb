# Necessary Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from sklearn import linear_model
from pandas.tools.plotting import scatter_matrix
import statsmodels.api as sm
import statsmodels.stats.api as sms
import statsmodels.formula.api as ols
from sklearn import preprocessing
matplotlib.style.use('ggplot')
import math
```

# Reading from the csv file

```python
car_data =  pd.read_csv('Ex04_Mpg_Regression.csv')
```

```python
car_data['displacement']
```

# Data Cleaning

```python
car_data = car_data.replace('?', np.NaN)
car_data.horsepower = car_data.horsepower.astype('float64')
car_data['carname'] = pd.Categorical(car_data.carname).codes
car_data = car_data.dropna(axis=0)
names = car_data.columns.tolist()
#car_data['displacement'] = car_data.displacement.apply(lambda x:1./(x + 0.0000001))
#car_data['weight'] = car_data.weight.apply(lambda x:1./(x + 0.0000001))
#car_data['horsepower'] = car_data.horsepower.apply(lambda x:1./(x + 0.000001))
car_data
```

# Normalizing the data

```python
x = car_data.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(car_data)
car_data = pd.DataFrame(x_scaled)
car_data.columns = names
car_data
```

# Finding the correaltion amongst variable

```python
correlation_before_feature_selection =  car_data.corr()
```

# Feature Plot (Correlation Matrix)

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlation_before_feature_selection, vmin = -1, vmax = 1)
fig.colorbar(cax)
ticks = np.arange(0, len(names), 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

# Feature wise Plot

```python
scatter_matrix(car_data, alpha=0.2, figsize=(10, 10), diagonal='kde')
plt.show()
```

# Shuffling and splitting data (80/20)

```python
splitend = int(0.8 * len(car_data))
car_data_shuffled = car_data.sample(frac=1)
train, test = car_data_shuffled[:splitend], car_data_shuffled[splitend:]
train.columns
```

# Creating training and testing data

```python
X_train = train.iloc[:, [4,6,7]]
Y_train = train.iloc[:, 0] 
X_test = test.iloc[:, [4,6,7]]
Y_test = test.iloc[:, 0]
```

# Creating a linear regression model

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)
plt.scatter(list(Y_test), list(regr.predict(X_test)), color='Red')
plt.show()
```

# Printing Mean Squared error and Variance

```python
print("Coefficient:\n",regr.coef_)
print("Mean squared error: %.2f"
      % np.mean((regr.predict(X_test) - Y_test) ** 2))
print('Variance score: %.2f' % regr.score(X_test, Y_test))
```

# Plotting OLS Model

```python
mpg_cylinder = car_data.groupby(['cylinders', 'mpg']).count()
lst = mpg_cylinder.columns.tolist()
mpg_cylinder.drop(lst,axis = 1,inplace = True)
mpg_cylinder.reset_index(inplace = True)
plt.scatter(mpg_cylinder['cylinders'],mpg_cylinder['mpg'])
plt.show()
```

```python
mpg_displacement = car_data.groupby(['displacement', 'mpg']).count()
lst = mpg_displacement.columns.tolist()
mpg_displacement.drop(lst,axis = 1,inplace = True)
mpg_displacement.reset_index(inplace = True)
plt.scatter(mpg_displacement['displacement'],mpg_displacement['mpg'])
plt.show()
```

```python
mpg_modelyear = car_data.groupby(['modelyear', 'mpg']).count()
lst = mpg_modelyear.columns.tolist()
mpg_modelyear.drop(lst,axis = 1,inplace = True)
mpg_modelyear.reset_index(inplace = True)
plt.scatter(mpg_modelyear['modelyear'],mpg_modelyear['mpg'])
plt.show()
```

```python
mpg_weight = car_data.groupby(['weight', 'mpg']).count()
lst = mpg_weight.columns.tolist()
mpg_weight.drop(lst,axis = 1,inplace = True)
mpg_weight.reset_index(inplace = True)
plt.scatter(mpg_weight['weight'],mpg_weight['mpg'])
plt.show()
```

```python
mpg_acceleration = car_data.groupby(['acceleration', 'mpg']).count()
lst = mpg_acceleration.columns.tolist()
mpg_acceleration.drop(lst,axis = 1,inplace = True)
mpg_acceleration.reset_index(inplace = True)
plt.scatter(mpg_acceleration['acceleration'],mpg_acceleration['mpg'])
plt.show()
```

```python
model1 = ols.ols(formula='mpg ~cylinders+modelyear + origin+carname+weight+displacement', data=train).fit()
print (model1.summary())
res = model1.predict(X_test)
plt.scatter(list(Y_test), list(res), color='Red')
plt.xlim(0.0,0.7)
plt.ylim(0.0,0.7)
plt.show()
```
