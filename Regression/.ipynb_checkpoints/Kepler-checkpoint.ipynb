```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from sklearn import linear_model
from pandas.tools.plotting import scatter_matrix
import statsmodels.api as sm
import statsmodels.stats.api as sms
import statsmodels.formula.api as ols
from sklearn import preprocessing
matplotlib.style.use('ggplot')
import math
```

```python
kepler_data =  pd.read_csv('Ex01_Kepler_Regression.csv')
kepler_data.planet = pd.Categorical(kepler_data.planet).codes
```

```python
names = kepler_data.columns.tolist()
x = kepler_data.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(kepler_data)
kepler_data = pd.DataFrame(x_scaled)
kepler_data.columns = names
kepler_data.SemiMajorAxis = kepler_data.SemiMajorAxis.apply(lambda x: math.pow(x, 2))
```

```python
splitend = int(0.9 * len(kepler_data))
kepler_data_shuffled = kepler_data.sample(frac=1)
train, test = kepler_data_shuffled[:splitend], kepler_data_shuffled[splitend:]
```

```python
correlation_before_feature_selection =  kepler_data.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlation_before_feature_selection, vmin = -1, vmax = 1)
fig.colorbar(cax)
ticks = np.arange(0, len(names), 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```python
scatter_matrix(kepler_data, alpha=0.2, figsize=(10, 10), diagonal='kde')
plt.show()
```

```python
X_train = train.iloc[:, 0:5]
Y_train = train.iloc[:, 5] 
X_test = test.iloc[:, 0:5]
Y_test = test.iloc[:, 5]
```

```python
model_scaled = ols.ols(formula='OrbitalPeriod ~ SemiMajorAxis', data=train).fit()
print (model_scaled.summary()) 
res = model_scaled.predict(X_test)
plt.plot(list(Y_test), list(res), color='Red')
plt.show()
```

```python
residual = res - Y_test
residual.plot(kind= 'box')
plt.show()
```

```python
np.sqrt(np.mean(residual ** 2))
```
